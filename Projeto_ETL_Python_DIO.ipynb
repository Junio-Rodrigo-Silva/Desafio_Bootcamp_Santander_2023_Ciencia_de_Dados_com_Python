```python
#Importando a biblioteca Pandas
import pandas as pd
```

#<font color=cyan>**EXTRACT**

```python
#Importanto a base
dados = pd.read_csv('/escolas.csv')

#Verificando os dados do arquivo
dados
```

#<font color=yellow>**TRANSFORM**

```python
#Transformando a coluna Orcamento_Anual para moeda (R$)
def incluir_moeda(real):
  return "R${:,.2f}".format(real)

dados["Orcamento_Anual"] = dados["Orcamento_Anual"].apply(incluir_moeda)
dados
```

#<font color=green>**LOAD**

```python
#Salvando o arquivo com os dados alterados
dados.to_csv("orcamento_escolar_formatodo_em_real.csv", index=False)
```
